```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab/parsernaam/data/fl_reg_data.csv")

df.head()
```

```python
df = df[['name_first','name_last']]
df.head()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.reset_index(drop=True, inplace=True)
df['name_first'] = df['name_first'].str.replace("[^a-zA-Z' -]", '', regex=True)
df['name_last'] = df['name_last'].str.replace("[^a-zA-Z' -]", '', regex=True)

df['name_first'] = df.name_first.str.strip().str.title()
df['name_last'] = df.name_last.str.strip().str.title()
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)
mask = np.random.rand(len(df)) < 0.5

df['name'] = np.where(mask, df['name_first'] + ' ' + df['name_last'], df['name_last'] + ' ' + df['name_first'])
df['type'] = np.where(mask, 'first_last', 'last_first')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

train_df, rest_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['type'])
val_df, test_df = train_test_split(df, test_size=0.5, random_state=42, stratify=df['type'])
```

```python
print(train_df.shape)
print(val_df.shape)
print(test_df.shape)
```

```python
train_df.to_csv('/content/drive/MyDrive/Colab/parsernaam/data/pos_train.csv', index=False)
val_df.to_csv('/content/drive/MyDrive/Colab/parsernaam/data/pos_val.csv', index=False)
test_df.to_csv('/content/drive/MyDrive/Colab/parsernaam/data/pos_test.csv', index=False)
```
